```python
import random


tasks = {
1 : "WD-evening",
2 : "WD-afternoon",
3 : "WE-afternoon",
4 : "WE-morning",
5 : "WE-evening",
6 : "WD-morning",
7 : "WE-anytime",
8 : "WD-anytime",
9 : "WD-night",
10 : "WE-night",
}


times_solts = len(tasks)
```

```python

def generate_chromosome():
  random_randints = []
  for task in tasks.keys() :
    random_randint = random.randint(1, times_solts)
    random_randints.append(random_randint)
  return random_randints

def fitness(chromosome):
    conflicts = len(chromosome) - len(set(chromosome))
    conflicts_value = 1 / (1 + conflicts)
    return conflicts_value


def selection(populations):
    weights = []
    for population in populations :
      weights.append(fitness(population))

    random_choices = random.choices(
        populations,
        weights=weights,
        k=2)

    return random_choices

def crossover(parent1, parent2):
    point = random.randint(1, times_solts-1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

def mutate(chromosome, rate):
    for i in range(len(chromosome)):
        if random.random() < rate:
            chromosome[i] = random.randint(1, times_solts)
    return chromosome

def genetic_algorithm(generations, population_size):
    population = []
    for i in range(population_size) :
      population.append(generate_chromosome())

    for generation in range(generations):
        new_population = []
        for i in range(population_size // 2):
            parent1, parent2 = selection(population)
            child1, child2 = crossover(parent1, parent2)
            new_population.append(mutate(child1,0.1))
            new_population.append(mutate(child2,0.1))
        population = new_population

        best = max(population, key=fitness)
        best_solution = []
        for i in best :
          best_solution.append(tasks.get(i))
        print(f"Generation {generation+1}: Best fitness = {fitness(best):.3f}, Solution : \n {best_solution} \n \n")
        if fitness(best) == 1 :
          break;
    return max(population, key=fitness)


```

```python

best_solution = genetic_algorithm(50,15)
print( best_solution)
```
